```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the results produced by LinearRegression.py
results_df = pd.read_csv('C:/USD/ms-aai-501-final-project/linear_regression_results.csv')

# Extract actual and predicted values
y_test = results_df['Actual']
y_predictions = results_df['Predicted']

# Plotting Actual vs. Predicted values
plt.figure(figsize=(10,6))
plt.scatter(y_test, y_predictions, alpha=0.5)
plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=4)
plt.show()

# Optionally, plotting residuals
residuals = y_test - y_predictions
plt.figure(figsize=(10,6))
plt.scatter(y_predictions, residuals, alpha=0.5)
plt.title('Residuals of Predictions')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.axhline(y=0, color='k', linestyle='--')
plt.show()
```
